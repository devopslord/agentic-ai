```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Scientific judge: every model ranks all answers; final result = average of rankings.
# Uses local Ollama (OPENAI_API_BASE, OPENAI_API_KEY from .env). Each model in competitors is asked to rank.

import re

def parse_judge_response(text):
    """Extract JSON {\"results\": [\"1\", \"2\", ...]} from model output. Returns list of ranks (1st, 2nd, ... as competitor numbers)."""
    text = text.strip()
    # Try to find JSON in the response (may be wrapped in markdown/code blocks)
    match = re.search(r'\{[^{}]*"results"\s*:\s*\[[^\]]*\]\s*\}', text)
    if match:
        try:
            data = json.loads(match.group())
            return [str(x) for x in data["results"]]
        except (json.JSONDecodeError, KeyError):
            pass
    try:
        data = json.loads(text)
        return [str(x) for x in data["results"]]
    except (json.JSONDecodeError, KeyError):
        return None

def ranking_to_ranks_per_competitor(order_list, n):
    """Convert [\"2\",\"1\",\"4\",\"3\"] (best to worst) to ranks: competitor i gets rank r. Returns list of length n."""
    ranks = [0] * n
    for position, comp_num in enumerate(order_list):
        idx = int(comp_num) - 1
        if 0 <= idx < n:
            ranks[idx] = position + 1  # 1-based rank
    return ranks

openai = OpenAI(
    base_url=os.getenv("OPENAI_API_BASE", "http://localhost:11434/v1"),
    api_key=os.getenv("OPENAI_API_KEY", "sk-local"),
)
n = len(competitors)
all_rankings = []  # list of lists: each inner list is rank per competitor from one judge

for model_name in competitors:
    try:
        response = openai.chat.completions.create(
            model=model_name,
            messages=judge_messages,
        )
        text = response.choices[0].message.content
        order = parse_judge_response(text)
        if order and len(order) == n:
            ranks = ranking_to_ranks_per_competitor(order, n)
            all_rankings.append(ranks)
            print(f"Judge {model_name}: {order}")
        else:
            print(f"Judge {model_name}: (parse failed, skipping)")
    except Exception as e:
        print(f"Judge {model_name}: ({e})")

# Average rank per competitor (lower = better)
import statistics
if not all_rankings:
    raise ValueError("No model produced a valid ranking. Check judge prompt and model outputs.")
avg_ranks = [statistics.mean([r[i] for r in all_rankings]) for i in range(n)]
# Sort by average rank, then by competitor index for ties
final_order = sorted(range(n), key=lambda i: (avg_ranks[i], i))
results = json.dumps({"results": [str(comp_idx + 1) for comp_idx in final_order]})  # compatibility with next cell
print("\n--- Average ranking (final) ---")
for rank_idx, comp_idx in enumerate(final_order):
    print(f"Rank {rank_idx+1}: {competitors[comp_idx]} (avg rank: {avg_ranks[comp_idx]:.2f})")
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```
